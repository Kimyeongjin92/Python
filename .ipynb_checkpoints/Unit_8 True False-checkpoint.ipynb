```python
# [Unit8: 불과 비교, 논리 연산자]

### 1. 비교연산자 ------------------------------------------------------------

###(숫자)###
print(3>1,        # 3이 1보다 크다.
      10==10,     # 10과 10은 같다   
      10 != 5,    # 10과 5는 다르다
      10 != 10,   # 10과 10은 다르다
      10 >= 20, 
      10 <= 20,   # 10이 20보다 작거나 같다.
      sep='  ')

###(문자)###
print('\n',
      'python' == 'python', # 문자열끼리 비교.
      'R_pro'  == 'python')


###(객체)###
print('\n',
      1 == 1.0,     # 값   vs 값   (값이 같은가 - 그래)
      1 is 1.0,     # 정수 vs 실수 (객체가 같은가 - 아니)
      1 is not 1.0  # 정수 vs 실수 (객체가 다른가 - 그래)
     )


### 2. 논리연산자 (and or not)  -------------------------------------------------

### and 두 값이 모두 True면 True
print('\n',
      True and True,
      True and False,
     False and False)

### or 한 값이라도 True면 True
print('\n',
      True or True,
      True or False,
      False or False)

### not  논리값을 뒤집는다.
print('\n',
      not True,
      not False,
      not True and False or not False,
      False and False or True,
      False or True)


### 3. 비교 & 논리 연산자 --------------------------------------------------------

print('\n',
     10 == 10 and 10 != 5,
     10 >5 or 10 <3,
     not 10 > 5,
     not 1 is 1.0)


#=================================== 연습문제 ====================================
korean = 92
english = 27
mathematics = 86
science = 81

print('\n',
      korean >= 50 and
      english >= 50 and
      mathematics >= 50 and 
      science >= 50)
```
